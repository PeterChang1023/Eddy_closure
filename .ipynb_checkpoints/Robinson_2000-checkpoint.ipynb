```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.sparse import diags
from scipy.sparse.linalg import spsolve

# === Grid and constants ===
ny, nz_new = 301, 401
Ly = 7500e3
p_bot, p_top = 1000e2, 100e2
p_new = np.linspace(p_bot, p_top, nz_new)
y = np.linspace(-Ly, Ly, ny)
dy = y[1] - y[0]
dp_new = p_new[1] - p_new[0]
Y, P_new = np.meshgrid(y, p_new, indexing='ij')

# Physical constants
f0 = 1e-4
sigma = 1e-5
tau_rad = 40 * 86400
tau_ek = 6.5 * 86400
pb = p_bot
pt = 200e2
pl = 500e2
Ab = 1e-5 * (pb - pt)
width = 1500e3
y0 = -1500e3

# === Forcing ===
# Barotropic forcing [v*q]_b (Eq. 18)
vqb_b = Ab * np.exp(-(y / width)**2)


# Interior forcing (Eq. 20)
A_p = np.zeros_like(p_new)
mask = (p_new <= pl) & (p_new > pt)
A_p[mask] = -Ab / (pl - pt)

vqb_ext = np.zeros((ny, nz_new))
for j in range(nz_new):
    vqb_ext[:, j] = A_p[j] * np.exp(-((y - y0) / width)**2)

# === Laplacian operator in y ===
diag = -2 * np.ones(ny)
off1 = np.ones(ny - 1)
Laplacian = diags([off1, diag, off1], [-1, 0, 1]) / dy**2
Laplacian = Laplacian.tolil()
Laplacian[0, 0] = Laplacian[0, -1] = 1
Laplacian[-1, -1] = Laplacian[-1, 0] = 1

# === Solve for barotropic wind u_b ===
rhs_ub = (tau_ek / pb) * np.gradient(np.gradient(vqb_b, dy), dy)
ub = spsolve(Laplacian.tocsc(), rhs_ub)
ub_2d = np.tile(ub[:, np.newaxis], (1, nz_new))

# === Solve Poisson equation for baroclinic response (Eq. 13) ===
prefac = f0**2 / (tau_rad * sigma)
rhs_full = np.gradient(np.gradient(vqb_ext, dy, axis=0), dy, axis=0) / prefac

u_jacobi_full = np.zeros((ny, nz_new))
max_iter = 5000
tolerance = 1e-8
j_t = np.argmax(p_new < pt)  # index for top of interior forcing

for it in range(max_iter):
    u_old = u_jacobi_full.copy()
    for j in range(1, nz_new - 1):
        u_jacobi_full[:, j] = 0.5 * (u_old[:, j - 1] + u_old[:, j + 1] - rhs_full[:, j] * dp_new**2)

    # Neumann BC at model top (100 hPa)
    u_jacobi_full[:, -1] = u_jacobi_full[:, -2]

    # Barotropic constraint above forcing (p < pt)
    u_jacobi_full[:, j_t+1:] = u_jacobi_full[:, j_t][:, np.newaxis]

    res = np.max(np.abs(u_jacobi_full - u_old))
    if res < tolerance:
        break

# === Combine baroclinic + barotropic ===
u_total = u_jacobi_full + ub_2d

# === Plot ===
plt.figure(figsize=(8, 6))
cs = plt.contourf(y / 1e3, p_new / 100, u_total.T, levels=21, cmap="RdBu_r")
plt.colorbar(cs, label='Total Zonal Wind [m/s]')
plt.contour(y / 1e3, p_new / 100, u_total.T, levels=10, colors='k', linewidths=0.5)
plt.xlabel('y (km)')
plt.ylabel('Pressure (hPa)')
plt.title('Total Zonal Wind $u(y,p)$ (Barotropic + Baroclinic)')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# Re-run after kernel reset
import numpy as np
import matplotlib.pyplot as plt

# === Grid and constants ===
ny, nz = 101, 101
Ly = 5000e3
p_top, p_bot = 0, 1000e2
y = np.linspace(-Ly, Ly, ny)
p = np.linspace(p_bot, p_top, nz)
dy, dp = y[1] - y[0], p[1] - p[0]
Y, P = np.meshgrid(y, p, indexing='ij')

f0 = 1e-4
sigma = 1e-5
R = 287
T_b = 250
width = 1500e3

# Forcing parameters
pb, pt = 1000e2, 200e2
Ab_bottom = 10 * (pb - pt) / 100 #/ 86400  # m/s/day
vqb_b = Ab_bottom * np.exp(-(y / width)**2)
d2vqb_dy2 = np.gradient(np.gradient(vqb_b, dy), dy)

# === Initial condition ===
u_sor = np.zeros_like(Y)

# === Boundary condition at bottom (Eq. 12) ===
u0 = np.zeros(ny)
for j in range(ny):
    coeff0 = -f0**2 / (sigma * dp)
    coeff1 = f0**2 * pb / (R * T_b)
    rhs0 = -d2vqb_dy2[j]
    u0[j] = rhs0 / (coeff1)
    u_sor[j, 0] = u0[j]

# === SOR solver ===
alpha = f0**2 / sigma
omega = 1.5
max_iter = 20000
tolerance = 1e-6
rhs = np.zeros_like(Y)

residuals = []

for iteration in range(max_iter):
    max_resid = 0.0
    for j in range(1, ny - 1):
        for k in range(1, nz - 1):
            d2udy2 = (u_sor[j+1, k] + u_sor[j-1, k]) / dy**2
            d2udp2 = (u_sor[j, k+1] + u_sor[j, k-1]) / dp**2
            denom = 2 / dy**2 + 2 * alpha / dp**2
            u_new = (rhs[j, k] + d2udy2 + alpha * d2udp2) / denom
            resid = u_new - u_sor[j, k]
            u_sor[j, k] += omega * resid
            max_resid = max(max_resid, abs(resid))

    # Enforce boundary conditions
    u_sor[:, 0] = u0
    u_sor[:, -1] = u_sor[:, -2]  # Neumann at top
    u_sor[0, :] = 0
    u_sor[-1, :] = 0

    residuals.append(max_resid)
    if max_resid < tolerance:
        break
```

```python
from scipy.sparse import lil_matrix
from scipy.sparse.linalg import spsolve

# === Redefine grid and constants ===
ny, nz = 101, 101
Ly = 5000e3
p_top, p_bot = 0, 1000e2
y = np.linspace(-Ly, Ly, ny)
p = np.linspace(p_bot, p_top, nz)
dy, dp = y[1] - y[0], p[1] - p[0]
Y, P = np.meshgrid(y, p, indexing='ij')

# Physical constants
f0 = 1e-4
sigma = 1e-5
R = 287
T_b = 280
pb = p[0]

# === Forcing ===
width = 1500e3
vqb_b = 1 * np.exp(-(y / width) ** 2)
d2vqb_dy2 = np.gradient(np.gradient(vqb_b, dy), dy)
rhs_bc = -d2vqb_dy2

# === Bottom BC: Solve Eq (12) with u1 = 0 ===
A = f0**2 / (sigma * dp)
B = f0**2 * pb / (R * T_b)
u1 = np.zeros(ny)
u0 = (rhs_bc + A * u1) / (A + B)

# === Initialize matrix system: A u = b ===
# Rebuild system with corrected RHS (forcing only at surface level)

# === Redefine matrix system: A u = b ===
N = ny * nz
A_mat = lil_matrix((N, N))
b_vec = np.zeros(N)

def idx(i, j):
    return j * ny + i

for j in range(nz):
    for i in range(ny):
        k = idx(i, j)

        # Bottom boundary (Dirichlet from ODE solution)
        if j == 0:
            A_mat[k, k] = 1
            b_vec[k] = u0[i]

        # Top Neumann BC: u[-1] = u[-2]
        elif j == nz - 1:
            A_mat[k, idx(i, j)] = 1
            A_mat[k, idx(i, j - 1)] = -1
            b_vec[k] = 0

        # Interior points
        elif 0 < j < nz - 1 and 0 < i < ny - 1:
            A_mat[k, idx(i - 1, j)] = 1 / dy**2
            A_mat[k, idx(i, j)] = -2 / dy**2 - 2 * f0**2 / (sigma * dp**2)
            A_mat[k, idx(i + 1, j)] = 1 / dy**2
            A_mat[k, idx(i, j - 1)] = f0**2 / (sigma * dp**2)
            A_mat[k, idx(i, j + 1)] = f0**2 / (sigma * dp**2)
            b_vec[k] = d2vqb_dy2[i] if j == 1 else 0  # forcing only just above bottom

        # y-boundaries: Dirichlet zero
        else:
            A_mat[k, k] = 1
            b_vec[k] = 0

# === Solve ===
u_flat = spsolve(A_mat.tocsr(), b_vec)
u_direct = u_flat.reshape((ny, nz))



```

```python
import numpy as np
import matplotlib.pyplot as plt

# === Grid and Constants ===
ny, nz = 101, 101
Ly = 5000e3
p_top, p_bot = 0, 1000e2
y = np.linspace(-Ly, Ly, ny)
p = np.linspace(p_bot, p_top, nz)
dy, dp = y[1] - y[0], p[1] - p[0]
Y, P = np.meshgrid(y, p, indexing='ij')

# === Physical constants ===
f0 = 1e-4
sigma = 1e-5
R = 287
T_b = 288
pb = p[0]

# === Forcing ===
width = 1500e3
Ab = (4000 / 86400) * (pb - 200e2)
vqb_b = Ab * np.exp(-(y / width)**2)
d2vqb_dy2 = np.gradient(np.gradient(vqb_b, dy), dy)
rhs_bc_corrected = -d2vqb_dy2

# === Coefficients ===
fcoeff = f0**2 / sigma
gamma = f0**2 * (1 / (sigma * dp) + pb / (R * T_b))

# === SOR Parameters ===
omega = 1.7
tolerance = 1e-4
max_iter = 5000

# === Initialization ===
u_sor = np.zeros((ny, nz))
rhs = np.zeros((ny, nz))
rhs[:, 1] = d2vqb_dy2  # forcing only at j = 1

# === SOR Loop ===
for it in range(max_iter):
    max_diff = 0

    # Update interior
    for j in range(1, nz - 1):
        for i in range(1, ny - 1):
            lap_y = (u_sor[i+1, j] + u_sor[i-1, j]) / dy**2
            lap_p = (u_sor[i, j+1] + u_sor[i, j-1]) / dp**2
            denom = -2 * fcoeff / dp**2 - 2 / dy**2
            rhs_term = rhs[i, j]
            u_new = (rhs_term - fcoeff * lap_p - lap_y) / denom
            delta = omega * (u_new - u_sor[i, j])
            u_sor[i, j] += delta
            max_diff = max(max_diff, abs(delta))

    # === Boundary Conditions ===

    # Bottom (Robin)
    for i in range(1, ny - 1):
        u_sor[i, 0] = u_sor[i, 1] - rhs_bc_corrected[i] / gamma

    # Top (Neumann)
    u_sor[:, -1] = u_sor[:, -2]

    # Meridional (Neumann)
    u_sor[0, :] = u_sor[1, :]
    u_sor[-1, :] = u_sor[-2, :]

    if max_diff < tolerance:
        print(f"SOR converged in {it} iterations.")
        break
else:
    print("SOR did not converge.")


```

```python
import numpy as np
import matplotlib.pyplot as plt

# === Grid and Constants ===
ny, nz = 101, 101
Ly = 5000e3
p_top, p_bot = 0, 1000e2
y = np.linspace(-Ly, Ly, ny)
p = np.linspace(p_bot, p_top, nz)  # bottom to top
dy, dp = y[1] - y[0], p[1] - p[0]
Y, P = np.meshgrid(y, p, indexing='ij')

# Physical constants
f0 = 1e-4
sigma = 1e-5
R = 287
T_b = 288
pb = p[0]

# === Forcing: Eddy PV flux divergence at the bottom ===
width = 1500e3
Ab = (100) * (pb - 200e2)   # as in Robinson (2000)
vqb_b = Ab * np.exp(-(y / width)**2)
d2vqb_dy2 = np.gradient(np.gradient(vqb_b, dy), dy)
rhs_bc = -d2vqb_dy2

# === Coefficients for BC ===
gamma_bc = f0**2 / sigma
alpha_bc = f0**2 * pb / (R * T_b)

# === Solve for u_0 using Robin boundary condition ===
u0 = np.zeros(ny)
u1 = np.zeros(ny)
for i in range(1, ny - 1):
    denom = gamma_bc / dp + alpha_bc
    u0[i] = rhs_bc[i] / denom
    dudp_bot = (rhs_bc[i] - alpha_bc * u0[i]) / gamma_bc
    u1[i] = u0[i] + dp * dudp_bot

# === Initialize full field ===
u_sor = np.zeros((ny, nz))
u_sor[:, 0] = u0
u_sor[:, 1] = u1

# === RHS forcing: only in second pressure level (j=1) ===
rhs = np.zeros((ny, nz))
rhs[:, 0] = d2vqb_dy2

# === SOR Parameters ===
fcoeff = f0**2 / sigma
omega = 1.7
tolerance = 1e-4
max_iter = 5000

# === SOR Iteration ===
for it in range(max_iter):
    max_diff = 0
    for j in range(2, nz - 1):  # Start from j=2 since j=0,1 are initialized
        for i in range(1, ny - 1):
            lap_y = (u_sor[i + 1, j] + u_sor[i - 1, j]) / dy**2
            lap_p = (u_sor[i, j + 1] + u_sor[i, j - 1]) / dp**2
            rhs_term = rhs[i, j]
            denom = -2 * fcoeff / dp**2 - 2 / dy**2
            u_new = (rhs_term - fcoeff * lap_p - lap_y) / denom
            delta = omega * (u_new - u_sor[i, j])
            u_sor[i, j] += delta
            max_diff = max(max_diff, abs(delta))

    # === Boundary Conditions ===
    u_sor[:, -1] = u_sor[:, -2]         # Top boundary: Neumann (∂u/∂p = 0)
    u_sor[0, :] = u_sor[1, :]           # y-boundary: ∂u/∂y = 0
    u_sor[-1, :] = u_sor[-2, :]

    if max_diff < tolerance:
        break

print(it)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# === Grid and Constants ===
ny, nz = 101, 101
Ly = 5000e3
p_top, p_bot = 0, 1000e2
y = np.linspace(-Ly, Ly, ny)
p = np.linspace(p_bot, p_top, nz)
dy, dp = y[1] - y[0], p[1] - p[0]
Y, P = np.meshgrid(y, p, indexing='ij')

# === Physical constants ===
f0 = 1e-4
sigma = 1e-5
R = 287
T_b = 288
pb = p[0]

# === Forcing ===
width = 1500e3
Ab = (40000 / 86400) * (1000 - 200)  # m²/s²/day × hPa
vqb_b = Ab * np.exp(-(y / width)**2)
d2vqb_dy2 = np.gradient(np.gradient(vqb_b, dy), dy)

# === Coefficients ===
fcoeff = f0**2 / sigma
rhs_bc_corrected = -d2vqb_dy2

# === SOR Settings ===
omega = 1.7
tolerance = 1e-4
max_iter = 5000

# === Initial Guess ===
u_sor = np.zeros((ny, nz))

# === RHS forcing: ONLY at bottom (baroclinic) level ===
rhs = np.zeros((ny, nz))
rhs[:, 1] = d2vqb_dy2


# === SOR Iteration ===
for it in range(max_iter):
    max_diff = 0
    # Top BC: Neumann
    u_sor[:, -1] = u_sor[:, -2]

    # Meridional BCs: Neumann
    u_sor[0, :] = u_sor[1, :]
    u_sor[-1, :] = u_sor[-2, :]


    # Bottom BC via Eq (12)
    denom_bc = 1 - (sigma * pb * dp) / (R * T_b)
    numer_bc = u_sor[:, 1] + (sigma * dp / f0**2) * d2vqb_dy2
    u_sor[:, 0] = numer_bc / denom_bc

    # Interior update (skip boundaries)
    for i in range(1, ny - 1):
        for j in range(1, nz - 1):
            lap_y = (u_sor[i + 1, j] + u_sor[i - 1, j]) / dy**2
            lap_p = (u_sor[i, j + 1] + u_sor[i, j - 1]) / dp**2
            rhs_term = rhs[i, j]
            denom = -2 * fcoeff / dp**2 - 2 / dy**2
            u_new = (rhs_term - fcoeff * lap_p - lap_y) / denom
            delta = omega * (u_new - u_sor[i, j])
            u_sor[i, j] += delta
            max_diff = max(max_diff, abs(delta))




    if it % 100 == 0:
        print(f"Iteration {it}, max update = {max_diff:.3e}")

    if max_diff < tolerance:
        print(f"SOR converged in {it} iterations.")
        break
else:
    print("SOR did not converge.")


```

```python
import numpy as np
import matplotlib.pyplot as plt

# === Grid and constants ===
ny, nz = 51, 51
Ly = 5000e3
pb, pt = 1000e2, 200e2
y = np.linspace(-Ly, Ly, ny)
p = np.linspace(pb, 0, nz)
dy, dp = y[1] - y[0], (p[1] - p[0])
Y, P = np.meshgrid(y, p, indexing='ij')

f0 = 1e-4
sigma = 1e-5
R = 287
T_b = 288 
width = 1500e3

# === Forcing: bottom forcing only ===
Ab = 700 * (pb - pt)/100  # units: m^2/s^2 (approx 800)
vqb_b = Ab * np.exp(-(y / width)**2)
d2vqb_dy2 = np.gradient(np.gradient(vqb_b, dy), dy)

# === Constants for ODE ===
A = sigma * pb / (R * T_b)
B_vec = sigma / f0**2 * d2vqb_dy2

# === Initialization ===
u_t = np.zeros((ny, nz))
# u_t[:, 0] = 10  # or 10, per user preference
omega = 1.6
tolerance = 1e-8
max_iter = 2000

rhs = np.zeros((ny, nz))
# rhs[:, 0] = d2vqb_dy2

# u_t[:, 0] = u_t[:,1] * np.exp(A * dp) + B_vec * dp * np.exp(2 * A * dp)

# === SOR loop ===
for it in range(max_iter):
    u_old = u_t.copy()

    # Step 1: update j=1 using ODE from j=0
    # u_t[:, 1] = -B_vec * dp * np.exp(A * dp) + u_t[:, 0] * np.exp(-A * dp)
    u_t[:, 0] = B_vec * dp * np.exp(2* A * dp) + u_t[:, 1] * np.exp(A * dp)
    
    # Boundary conditions
    u_t[:, -1] = u_t[:, -2]     # ∂u/∂p = 0 at top
    u_t[0, :] = u_t[1, :]       # ∂u/∂y = 0 at left
    u_t[-1, :] = u_t[-2, :]     # ∂u/∂y = 0 at right
    
    # Step 2: update interior (from j=1 to nz-2)
    for j in range(1, nz - 1):
        for i in range(1, ny - 1):
            # d2udy2 = (u_t[i + 1, j] - 2 * u_t[i, j] + u_t[i - 1, j]) / dy**2
            # d2udp2 = (u_t[i, j + 1] - 2 * u_t[i, j] + u_t[i, j - 1]) / dp**2
            d2udy2 = (u_t[i + 1, j] + u_t[i - 1, j]) / dy**2
            d2udp2 = (u_t[i, j + 1] + u_t[i, j - 1]) / dp**2
            rhs_term = rhs[i,j] #0.0
            laplace_u = d2udy2 + (f0**2 / sigma) * d2udp2
            coeff = 2 / dy**2 + 2 * f0**2 / (sigma * dp**2)            
            u_new = (laplace_u - rhs_term) / coeff
            u_t[i, j] = (1 - omega) * u_t[i, j] + omega * u_new    

    if it % 100 == 0:
        print(f"Iteration {it}, max update = {np.max(np.abs(u_t - u_old)):.3e}")


    if np.max(np.abs(u_t - u_old)) < tolerance:
        break

```
